### Hadleigh's Tephigram example

#### 21st Jan 2016

#### As per Thompkins fig 1.19

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

temps = np.linspace(200.,350., 100., endpoint=True)
theta = np.linspace(200.,500., 100., endpoint=True)
tephi_temp, tephi_theta = np.meshgrid(temps, theta)

re_press = 1000
coeff = 1 / (287/1005)

def pressure(tephi_temp, tephi_theta):
    return ref_press * ((tephi_theta/tephi_temp)**(-coeff)) / 100

cp = plt.contour(tephi_temp, tephi_theta, pressure(tephi_temp, tephi_theta), \
                 levels=range(600,1200,200), colors='black', linewidth=.5)
plt.title('Hadleighs Tephigram')
plt.xlabel('Temp (K)')
plt.ylabel('Log Theta (K)')
plt.clabel(cp, inline=True)
plt.yscale('log')
plt.xticks()
plt.yticks()
plt.show()
```
